```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
init_signal = np.concatenate((np.zeros((1, 30)),
                              np.ones((1, 2)),
                              np.zeros((1, 20)),
                              np.ones((1, 30)),
                              2 * np.ones((1, 10)),
                              np.zeros((1, 30)),
                              -np.ones((1, 10)),
                              np.zeros((1, 40))), axis=1)[0]

kernel = np.exp(-np.linspace(-2, 2, 20) ** 2)
kernel /= sum(kernel)

convolved_signal = np.convolve(init_signal, kernel, 'same')

# Plotting
plt.figure(figsize=(15, 4))
plt.plot(kernel)
plt.xlim([0, len(init_signal)])
plt.title('Kernel')
plt.show()

plt.figure(figsize=(15, 4))
plt.plot(init_signal)
plt.xlim([0, len(init_signal)])
plt.title('Initial signal')
plt.show()

plt.figure(figsize=(15, 4))
plt.plot(convolved_signal)
plt.xlim([0, len(init_signal)])
plt.title('Convolved signal')
plt.show()
```

```python
kernel_sharp = np.exp(-np.linspace(-2, 2, 20) ** 2/0.05) # <----- denominator sharpens the filter and signal
kernel /= sum(kernel)

convolved_signal_sharp = np.convolve(init_signal, kernel_sharp, 'same')

# Plotting
plt.figure(figsize=(15, 4))
plt.plot(kernel_sharp)
plt.xlim([0, len(init_signal)])
plt.title('Kernel')
plt.show()

plt.figure(figsize=(15, 4))
plt.plot(init_signal)
plt.xlim([0, len(init_signal)])
plt.title('Initial signal')
plt.show()

plt.figure(figsize=(15, 4))
plt.plot(convolved_signal_sharp)
plt.xlim([0, len(init_signal)])
plt.title('Convolved signal (sharper)')
plt.show()
```
