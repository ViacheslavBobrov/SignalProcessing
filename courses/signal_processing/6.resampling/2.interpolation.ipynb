```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy import interpolate
```

```python
sampling_frequency = 10

signal = np.array([1, 4, 3, 6, 2, 19])

n_signal = len(signal)
time = np.arange(0, n_signal) / sampling_frequency

plt.plot(time, signal, 'ks-')
```

```python
interpolated_n_signal = 47
interpolated_time = np.linspace(time[0], time[-1], interpolated_n_signal)
# different interpolation methods
interpolation_methods = ['linear', 'nearest', 'cubic']

for i in range(len(interpolation_methods)):
    interpolated_signal = interpolate.griddata(time, signal, interpolated_time,
                                               method=interpolation_methods[i])
    plt.figure(figsize=(15, 7))
    plt.plot(interpolated_time, interpolated_signal, 'bs-',
             label=interpolation_methods[i] + ' interpolation')
    plt.plot(time, signal, 'rs-', label='initial signal')
    plt.legend()
    plt.show()
```
