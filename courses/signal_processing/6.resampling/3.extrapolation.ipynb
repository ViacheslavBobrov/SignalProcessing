```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy import interpolate
```

```python
sampling_frequency = 10

signal = np.array([1, 4, 3, 6, 2, 19])

n_signal = len(signal)
time = np.arange(0, n_signal) 

plt.plot(time, signal, 'ks-')
```

```python
extrapolation_time = np.arange(time[0] - 2, time[-1] + 3)
extrapolation_methods = ['linear', 'cubic']

for i in range(len(extrapolation_methods)):
    extrapolated_object = interpolate.interp1d(time, signal,
                                               kind=extrapolation_methods[i], fill_value='extrapolate')
    extrapolated_signal = extrapolated_object(extrapolation_time)
    plt.figure(figsize=(15, 7))
    plt.plot(extrapolation_time, extrapolated_signal, 'bs-',
             label=extrapolation_methods[i] + ' extrapolation')
    plt.plot(time, signal, 'rs-', label='initial signal')
    plt.ylim([-100, 100])
    plt.legend()
    plt.show()
```
