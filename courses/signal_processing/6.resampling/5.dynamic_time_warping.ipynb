```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean
```

```python
## create signals

# different time vectors
time_a = np.linspace(0, 1.5 * np.pi, 400)
time_b = np.linspace(0, 8 * np.pi, 100)

signal_a = np.sin(time_a ** 2)
signal_b = np.sin(time_b)

plt.figure(figsize=(15, 7))
plt.plot(time_a, signal_a, 'bs-', label='signal a')
plt.plot(time_b, signal_b, 'rs-', label='signal b')
plt.xlabel('Time (rad.)')
plt.legend()
plt.title('Before time warping')
```

```python
distance, path = fastdtw(signal_a, signal_b, dist=euclidean)
indices = np.asarray([[path[i][0], path[i][1]] for i in range(len(path))])
plt.figure(figsize=(15, 7))
plt.plot(signal_a[indices[:, 0]], 'b-', label='signal a')
plt.plot(signal_b[indices[:, 1]], 'r-', label='signal b')
plt.title('After time warping (signal b is changed to fit signal a)')
plt.legend()
```
